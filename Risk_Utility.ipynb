```python
import numpy as np
import matplotlib.pyplot as plt

# Utility function
def utility(x, R):
    return 1 - np.exp(-x/R)

# Decision values and probabilities
decision_values = [68915327, 32579184]
probabilities = [0.8, 0.2]

# Reference point
R = 30000000
EV = 80436586.2

# Calculate Expected Utility (EU)
EU = 0.8 * (1 - np.exp(-decision_values[0] / R)) + 0.2 * (1 - np.exp(-decision_values[1] / R))

# Certainty Equivalent (CE) calculation
CE = -R * np.log(1 - EU)

# Risk Premium (RP) calculation
RP = EV - CE

# Plotting
x_values = np.linspace(0, 2 * decision_values[0], 1000)
y_values = utility(x_values, R)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Risk Utility Curve')
plt.axhline(y=EU, color='r', linestyle='-', label='Expected Utility (EU)')
plt.axvline(x=CE, color='g', linestyle='--', label='Certainty Equivalent (CE)')
plt.plot([CE, EV], [EU - 0.05, EU - 0.05], color='orange', linestyle='--', label='Risk Premium') # Risk Premium line
plt.axvline(x=EV, ymin=0, ymax=1, color='purple', linestyle='--', label=f'Expected Value (EV)')  # EV line
plt.scatter([CE], [EU], color='g', label='Certainty Equivalent (CE)')

plt.xlabel('Wealth (x)')
plt.ylabel('Utility')
plt.title('Risk Utility Curve')
plt.legend()
plt.grid(True)

# Annotation for Risk Attitude
if CE > EV:
    risk_attitude = 'Risk-Seeking'
elif CE < EV:
    risk_attitude = 'Risk-Averse'
else:
    risk_attitude = 'Risk-Neutral'

plt.text(CE + 500, EU, f'CE ({CE:.2f})', fontsize=10, va='bottom', ha='left', color='blue')
plt.text(EV + 500, EU - 0.05, f'RP ({RP:.2f})', fontsize=10, va='top', ha='left', color='orange')
plt.text(CE + 500, EU - 0.1, risk_attitude, fontsize=10, va='top', ha='left', color='brown')

plt.show()

print(f'Expected Utility (EU): {EU:.2f}')
print(f'Certainty Equivalent (CE): {CE:.2f}')
print(f'Risk Premium (RP): {RP:.2f}')
print(f'Risk Attitude: {risk_attitude}')
```
