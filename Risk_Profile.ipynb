```python
import matplotlib.pyplot as plt

# Decision alternatives
decisions = ['Deploy continuous training & auditing team (s1)', 'Increase staff during peak season (s2)']

# Probabilities
probabilities = [0.8, 0.2]

# Profits in millions
profits = [51792364, 43875192]

# Plotting the risk profile graph
plt.figure(figsize=(10, 6))

# Plotting standing lines
for i, decision in enumerate(decisions):
    plt.plot([profits[i], profits[i]], [0, probabilities[i]], label=decision, linestyle='--', marker='o')

# Plotting the points
plt.scatter(profits, probabilities, color='blue')

# Adding labels and title
plt.xlabel('Profit (millions)')
plt.ylabel('Probabilities')
plt.title('Risk Profile Graph for Mitigate stolen bags (d1)')
plt.legend()

# Displaying the graph
plt.grid(True)
plt.show()
```
