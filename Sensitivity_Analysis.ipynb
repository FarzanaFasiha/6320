```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equations for each decision alternative
def ev_d1(p):
    return 51792364 * p + (43875192 * (1 - p))

def ev_d2(p):
    return 68915327 * p + (32579184 * (1 - p))

def ev_d3(p):
    return 96724813 * p + (15283679 * (1 - p))

# Define the range of values for p
p_values = np.linspace(0, 1, 100)

# Calculate expected values for each decision alternative
ev_values_d1 = ev_d1(p_values)
ev_values_d2 = ev_d2(p_values)
ev_values_d3 = ev_d3(p_values)

# Plot the results
plt.plot(p_values, ev_values_d1, label='Decision 1')
plt.plot(p_values, ev_values_d2, label='Decision 2')
plt.plot(p_values, ev_values_d3, label='Decision 3')

# Find the point of intersection
intersection_index = np.argwhere(np.diff(np.sign(ev_values_d1 - ev_values_d2)) != 0).reshape(-1)
intersection_point_p = p_values[intersection_index]

# Plot the point of intersection
plt.plot(intersection_point_p, ev_values_d1[intersection_index], 'ro')

# Label the point of intersection
plt.text(intersection_point_p + 0.01, ev_values_d1[intersection_index] + 1000000, f' (p={intersection_point_p[0]:.3f})', fontsize=10)

# Label axes and add legend
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value')
plt.title('Sensitivity Analysis')
plt.legend()
plt.grid(True)
plt.show()
```
